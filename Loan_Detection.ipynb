preprocessing and modeling

```python
from sklearn.compose import ColumnTransformer
from imblearn.pipeline import Pipeline
from sklearn.experimental import enable_iterative_imputer
from sklearn.impute import IterativeImputer ,SimpleImputer ,KNNImputer
from sklearn.metrics import confusion_matrix,f1_score
from sklearn.preprocessing import StandardScaler ,OneHotEncoder
from sklearn.model_selection import cross_validate
from imblearn.over_sampling import SMOTE
from imblearn.combine import SMOTETomek , SMOTEENN
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV
from mlxtend.feature_selection import SequentialFeatureSelector
import category_encoders as ce
```

```python
X = df.drop(["Loan_ID","Loan_Status"],axis=1)
y = df["Loan_Status"]
```

```python
models = {
    "DecisionTreeClassifier" : DecisionTreeClassifier(),
    "RandomForest":RandomForestClassifier(),
    "XGB":XGBClassifier(),
    "LGBM":LGBMClassifier(),
    "GaussianNB":GaussianNB(),
    "Knn":KNeighborsClassifier(),
    "svm" :SVC(),
}
```

```python
# Define the transformers separately
binary_encoder = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', ce.BinaryEncoder())
])

iterative_imputer = IterativeImputer(estimator=RandomForestClassifier())

# Define the column transformer
processor = ColumnTransformer(
    transformers=[
        ('binaryencoder', binary_encoder, ["Gender", "Married", "Education", "Self_Employed", "Property_Area"]),
        ('IterativeImputer', iterative_imputer, ["LoanAmount"]),
        ('KNNImputer', KNNImputer(n_neighbors=3, weights='uniform'), ["Dependents", "Credit_History", "Loan_Amount_Term"]),
    ],
    remainder='passthrough'
)
for model_name, model_instance in models.items():
    # Define the pipeline with the CustomMapper
    pipeline = Pipeline([
        ("TransformData", processor),
        ("scaling", StandardScaler()),
        ("model", model_instance)
    ])

    # Try to fit and evaluate the pipeline
    try:
        scores = cross_validate(pipeline, X, y, cv=5, scoring="f1", n_jobs=-1, return_train_score=True)
        print(model_name)
        print("Train F1-score:", scores["train_score"].mean())
        print("Test F1-score:", scores["test_score"].mean())
        print("_" * 25)
    except Exception as e:
        print(f"Error during pipeline execution: {e}")
```

```python
# Define the transformers separately
binary_encoder = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', ce.BinaryEncoder())
])

Onehot_encoder = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', ce.OneHotEncoder())
])

iterative_imputer = IterativeImputer(estimator=RandomForestClassifier())
# Define the column transformer
processor = ColumnTransformer(
    transformers=[
        ('IterativeImputer', iterative_imputer, ["LoanAmount"]),
        ('KNNImputer', KNNImputer(n_neighbors=3, weights='uniform'), ["Dependents", "Credit_History", "Loan_Amount_Term"]),
        ('binaryencoder', binary_encoder, ["Education", "Self_Employed", "Married"]),
        ("Onehotencoder",Onehot_encoder, ["Gender", "Property_Area"])
    ],
    remainder='passthrough'
)


for model_name, model_instance in models.items():
    # Define the pipeline with the CustomMapper
    pipeline = Pipeline([
        ("TransformData", processor),
        ("scaling", StandardScaler()),
        ("model", model_instance)
    ])

    # Try to fit and evaluate the pipeline
    try:
        scores = cross_validate(pipeline, X, y, cv=5, scoring="f1", n_jobs=-1, return_train_score=True, error_score='raise') 
        print(model_name)
        print("Train F1-score:", scores["train_score"].mean())
        print("Test F1-score:", scores["test_score"].mean())
        print("_" * 25)
    except Exception as e:
        print(f"Error during pipeline execution: {e}")
```

```python

for model_name, model_instance in models.items():
    # Define the pipeline with the CustomMapper
    pipeline = Pipeline([
        ("TransformData", processor),
        ("Undersampling",SMOTETomek()),
        ("scaling", StandardScaler()),
        ("model", model_instance)
    ])

    # Try to fit and evaluate the pipeline
    try:
        scores = cross_validate(pipeline, X, y, cv=5, scoring="f1", n_jobs=-1, return_train_score=True, error_score='raise') 
        print(model_name)
        print("Train F1-score:", scores["train_score"].mean())
        print("Test F1-score:", scores["test_score"].mean())
        print("_" * 25)
    except Exception as e:
        print(f"Error during pipeline execution: {e}")
```

```python
#oversampling

for model_name, model_instance in models.items():
    # Define the pipeline with the CustomMapper
    pipeline = Pipeline([
        ("TransformData", processor),
        ("Oversampling",SMOTEENN()),
        ("scaling", StandardScaler()),
        ("model", model_instance)
    ])

    # Try to fit and evaluate the pipeline
    try:
        scores = cross_validate(pipeline, X, y, cv=5, scoring="f1", n_jobs=-1, return_train_score=True, error_score='raise') 
        print(model_name)
        print("Train F1-score:", scores["train_score"].mean())
        print("Test F1-score:", scores["test_score"].mean())
        print("_" * 25)
    except Exception as e:
        print(f"Error during pipeline execution: {e}")
```

gridsearch

```python

# Define the pipeline
pipeline = Pipeline([
    ("TransformData", processor),
    ("scaling", StandardScaler()),
    ("model", SVC())  # Replace with your desired model, e.g., RandomForestClassifier()
])

# Define the parameter grid for GridSearchCV
param_grid = {
    'model__C': [0.1, 1, 10],
    'model__gamma': [0.01, 0.1, 1],
    'model__kernel': ['linear', 'rbf', 'poly'],
    'model__degree': [2, 3, 4],
}

# Create the GridSearchCV object
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='f1', n_jobs=-1, return_train_score=True)

```

```python
# Perform cross-validation
scores = cross_validate(grid_search, X, y, cv=5, scoring="f1", n_jobs=-1, return_train_score=True) 


print("Fit Time:", scores["fit_time"].mean())
print("Score Time:", scores["score_time"].mean())
print("Train F1-scores:", scores["train_score"].mean())
print("Test F1-scores:", scores["test_score"].mean())
print("Best Parameters:", grid_search.best_params_)
print("Best F1-score:", grid_search.best_score_)
print("_" * 25)
```
